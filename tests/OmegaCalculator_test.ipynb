## Cases for the different positions and sizes of Cherenkov radiation's cone and the telescope's reach and core.

* Geometric visualisation of the interaction between the Cherenkov cone and the telescope's mirror from Figure 13, Le Bohec et al, 1998, "A new analysis method for very high definition Imaging Atmospheric Cherenkov Telescopes as applied to the CAT telescope", Elsevier, pages 425 - 437




# The src class (a quick rundown):
## Functionalities
* update()
    - sets the parameters of the Cherenkov Cone and the Telescope in a 2D-plane.

* getOmegas()
    - returns  (angle between the intersection points between the cone and the telescope's annulus and/or core) and prints the intersection points should they exist.

* graphCircles()
    - plots a similar figure as in the bottom figure in Figure 13 from Le Bohec et al, 1998.
    - *NOTE*: This function only works when the axes origin is set to be at the centre of the Cherenkov radiation cone.

### Test results and analysis:

* the plotArc function called by graphCircles plots an arc in the counterclockwise direction. Thus for cases with only 1 pair of intersection points, it only works when the centre of the telescope is to the right of the cone.
    - Solution to make the plotArc function plot in the clockwise direction for this case has not been found.

* plotArc function has no issues with case 8 (2 pairs of intersection points).

* plotArc function should not be used when there are no intersection points.
