# Contextual knowledge

## Cases for the different positions and sizes of Cherenkov radiation's cone and the telescope's reach and core.

* Geometric visualisation of the interaction between the Cherenkov cone and the telescope's mirror from Figure 13, Le Bohec et al, 1998, "A new analysis method for very high definition Imaging Atmospheric Cherenkov Telescopes as applied to the CAT telescope", Elsevier, pages 425 - 437




# The src class (a quick rundown):
## Functionalities
* update()
    - sets the parameters of the Cherenkov Cone and the Telescope in a 2D-plane.

* getOmegas()
    - returns δω (angle between the intersection points between the cone and the telescope's annulus and/or core) and prints the intersection points should they exist.

* graphCircles()
    - plots a similar figure as in the bottom figure in Figure 13 from Le Bohec et al, 1998.
    - *NOTE*: This function only works when the axes origin is set to be at the centre of the Cherenkov radiation cone.


```python
# Import class
from src.omega_calculator import OmegaCalculator
```

```python
# Instantiate class
omega_calculator = OmegaCalculator()
```

```python
# Case 1
# Set new values
omega_calculator.update(cone_pos = (0, 0), tel_pos = (3, 3), r_cone = 1, r_tel_core = 1, r_tel_reach = 1.5)

# Solely graph the circles
omega_calculator.graphCircles()

# Calculate and print the omega
omega = omega_calculator.getOmegas()
print("Omega: ", omega, " degrees.")
```
