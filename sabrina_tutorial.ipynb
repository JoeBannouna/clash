```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar
from scipy.stats import rv_continuous
```

```python
class randgen(rv_continuous):
  def __init__(self, mu=0, sig=1, a=-10, b=10):
    super().__init__(self, a=a, b=b)
    self.mu = mu
    self.sig = sig

  def _pdf(self, x):
    return 1 / (self.sig * np.sqrt(2*np.pi)) * np.exp(-0.5 * (x-self.mu)**2 / self.sig**2)
```

```python
rg = randgen(mu=10)
plt.hist(rg.rvs(size=100))
plt.show()
```

```python
def func(x, a):
  xx = 10**x
  return 1e-4 * a*  xx**(-2)

def func2(x):
  return np.log2(1e-4) - 2*x

rs = np.random.rand(1000,2)

#rs[:,1] = 2*rs[:,1] - 15
rs = np.array([np.random.uniform(-2, 0, 1000), np.random.rand(1000)]).T
samples = np.array([])

for r in rs:
  if func(r[0]) > r[1]:
    samples = np.append(samples, 10**(r[0]))

plt.hist(samples)
plt.show()

print(len(samples))
```
